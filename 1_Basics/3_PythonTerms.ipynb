### Variables: The reference to an Object

**Object**: A defined name that references an object.

````python
job_title = "Data Analyst"
job_location = "US"
job_salary = 90000
````

```python
id()
```
Return the identity of an object.
This is guaranteed to be unique among simultaneously existing objects.

Python uses the object's memory address.

````python
id(job_title)
````

````python
job_1 = "DA"
job_2 = "DA"
id(job_1), id(job_2) # both the id's are different
````

````python
job_3 = job_1
id(job_3), id(job_1) # both the id's are same
````

### Classes: The Template of Objects
 **Classes**: A template for creating objects(records)


````python
job_title = "Data Analyst"
job_location = "US"
job_salary = 90000
````

````python
class jobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
````

````python
job_1 = jobPost(job_title, job_location, job_salary)
````

````python
job_1.location
````

### Methods: The Functions of an Object
 **Methods**: A functions defined inside a class that operates on its objects



````python
class jobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
  def display_info(self):
    print(f"Title:    {self.title}")
    print(f"Location: {self.location}")
    print(f"Salary:   {self.salary}")
````

````python
job_1 = jobPost(job_title, job_location, job_salary)
````

````python
job_1.title
````

````python
job_1.display_info()
````
