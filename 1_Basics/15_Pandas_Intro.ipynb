## Pandas: Cleaning

```python
import pandas as pd
from datasets import load_dataset

# load dataset
ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()
```

```python
type(df.job_posted_date[0])
```

```python
df.job_posted_date=pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df['job_posted_month']=df.job_posted_date.dt.month
```

```python
df['job_posted_month']
```

```python
df.info()
```

```python
df.sort_values('job_posted_date', inplace=True)
```

```python
df
```

```python
# axis - 0 for row and 1 for column
df.drop(labels='salary_hour_avg' , axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dropna(subset=['salary_year_avg'], inplace=True)
```

```python
df.info()
```

## Pandas: Analyzing

```python
import pandas as pd
from datasets import load_dataset

# loading dataset
ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.describe()
```

```python
df.info()
```

```python
df.count()
```

```python
df['salary_year_avg'].median()
```

```python
# min salary investigation
df['salary_year_avg'].min()
```

```python
# returns the index of the min value in df
min_sal=df['salary_year_avg'].idxmin()
```

```python
df.iloc[min_sal]
```

```python
# returns count of each row in a df

df['job_title_short'].value_counts()
```

```python
#
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
```

```python
df.groupby('job_title_short')[['salary_year_avg','salary_hour_avg']].median()
```

```python
# can aggregate groupby functions
df.groupby('job_title_short')['salary_year_avg'].agg(['min','max','median'])
```
