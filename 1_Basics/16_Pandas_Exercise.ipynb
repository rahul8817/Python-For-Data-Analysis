```python
# install dataset lib
# !pip install datasets

# import libraries
import pandas as pd
from datasets import load_dataset

# load data
ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

# cleanup data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country'].value_counts().head(20)
```

```python
# goes thru each column and checks if given iterable is in the column
df['job_country'].isin(['India']).any()
```

```python
ind_jobs = df[df['job_country'] == "India"]
```

```python
ind_jobs = ind_jobs[ind_jobs['salary_year_avg'].notna()]
```
