### **Python Built-in Data Types Reference**

| Category | Type | Example | Description |
| :--- | :--- | :--- | :--- |
| **Numeric** | `int` | `10`, `-5` | Whole numbers |
| | `float` | `10.5`, `3.14` | Decimal numbers |
| | `complex` | `1j`, `2 + 3j` | Imaginary components |
| **Sequence** | `str` | `"Hello"` | Character strings |
| | `list` | `[1, 2, 2]` | Ordered, **mutable** collection |
| | `tuple` | `(1, 2, 2)` | Ordered, **immutable** collection |
| | `range` | `range(5)` | Sequence of numbers |
| **Mapping** | `dict` | `{"a": 1}` | Key-value pairs |
| **Set** | `set` | `{1, 2, 3}` | Unordered, unique elements |
| | `frozenset` | `frozenset({1})` | Immutable version of set |
| **Boolean** | `bool` | `True`, `False` | Logical values |
| **Binary** | `bytes` | `b"Hi"` | Immutable byte sequences |
| **None** | `NoneType`| `None` | Represents "null" or no value |
