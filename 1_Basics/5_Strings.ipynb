### Concatenation

```python
role = "Data Analyst"
skill = "Python"
```

```python
'Role: ' + role
```

```python
help(str)
```

```python
# format(...)
# S.format(*args, **kwargs) -> str
# Return a formatted version of S, using substitutions from args and kwargs.
# The substitutions are identified by braces ('{' and '}').

role = "Data Analyst"
skill = "Python"

"Role: {}".format(role)
```

```python
# Create the string "Role: Data analyst; Skill: Python"
"Role: {}; Skill: {}".format(role, skill)
```

### f-Strings - Formatted String Liberals

```python
role = "Data Analyst"
skill = "Python"

f"Role: {role}; Skill: {skill}"
```

### %-formatting (printf-style string formatting)

```python
"Role: %s; Skill: %s" % (role,skill)
```
