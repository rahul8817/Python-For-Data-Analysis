# Exercise : Visualize Salaries

- calculating the median salary based on job title.
- distinguish b/w high and low paying roles

## Median Salary vs Data Science Jobs

```python
# install libraries

# import the libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load DS
ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

# cleanup data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind='barh')
plt.xlabel("Salary ($USD)")
plt.ylabel("Job Titles")
plt.title("Median Salary by Job Title")
plt.show()
```
