## Index Management


````python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
# cleanup data
df['job_posted_data'] = pd.to_datetime(df['job_posted_date'])
````

````python
df.sample(5)
````

```index.name``` and ```index.dtype```

````python
df.index.name = 'job_index' 
df.index
````

## 3 methods:

1. ```df.reset_index()```
2. ```df.set_index()```
3. ```df.sort_index()```

````python
df_usa = df[df['job_location'] == "United States"]

df_usa
````

````python
# reset_index()
df_usa.reset_index(inplace=True)
df_usa
````

````python
# set_index()

df_usa.set_index('job_index', inplace=True)
````

````python
df_usa.head()
````

````python
# sort_index()

median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
````

````python
median_pivot
````

````python
median_pivot.sort_values(by= ('median','salary_year_avg'), ascending=False)
````

````python
# sorted according to index , here it is sorted in alphabetical order
median_pivot.sort_index(inplace=True)
median_pivot
````
