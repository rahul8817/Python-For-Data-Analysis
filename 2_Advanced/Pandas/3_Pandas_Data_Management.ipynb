# Data Management:

- using ```df.sample()``` and ```df.copy()``` methods

### 1. ``` df.sample()```

````python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()  

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
````

````python
# returns different samples on every run, output varies all the time
df.sample(10)
````

````python
# random state fixes randomness, get the same result everytime you run the code
df.sample(10, random_state=42)
````

### 2. ```df.copy()```
- makes a copy of this object's indices and data 

````python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
````

````python
df_original
````

````python
df_altered = df_original
df_altered.loc[:5, 'salary_year_avg']
````

````python
median_salary = df_altered['salary_year_avg'].median()
median_salary
````

````python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)
df_altered.loc[:5, 'salary_year_avg']
````

````python
df_original.loc[:5, 'salary_year_avg']
````

````python
print("Original DF ID: ",id(df_original))
print("Altered DF ID: ", id(df_altered))
print("Are they both the same?: ",id(df_original) == id(df_altered))
````

- looking at the above codes, we can find out that when i tried to fill ```df_altered``` with the median values , ```df_original```'s values also got updated without even calling that df
- this can cause issues while working on data, so we can use the ```df.copy()``` method, where this fault wont occur 

````python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
````

````python
df_altered = df_original.copy()
````

````python
print("Original DF ID: ",id(df_original))
print("Altered DF ID: ", id(df_altered))
print("Are they both the same?: ",id(df_original) == id(df_altered))
````

````python
median_salary = df_altered['salary_year_avg'].median()
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)
df_altered.loc[:5, 'salary_year_avg']
````

````python
df_original.loc[:5, 'salary_year_avg']
````

- now when i used the ```copy()``` method, the ID's of both DFs are different meaning the original DF remains the same
- i can use this method which trying with work with df without changing it
