## Exercise - Investigate Trending Skills


```python
# import libraries - this time added ast lib, to cleanup the skills where it was previously as a str and not a list
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleanup data
df['job_skills'] = df['job_skills'].apply(
    lambda skill_list: ast.literal_eval(skill_list) 
    if pd.notna(skill_list) 
    else skill_list)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy() # use copy() method to avoid errors while making subsets
```

- need to agg the skills on a monthly basis, 

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month # created a new column showing the job posted month

# explode out the skills from the DF
df_DA_exploded = df_DA.explode('job_skills') 
```

```python
df_DA_exploded
# now we have individual job skills and the months

df_DA_pivot = df_DA_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

- creating an index called as total which contains the sum of each skill 

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
# need to sort the values , passing the indexes in a sorted manner to the pivot table
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

- now that we have the pivot, can drop the "Total" row

```python
df_DA_pivot = df_DA_pivot.drop('Total')
df_DA_pivot

```

```python
# plotting this table

df_DA_pivot.iloc[:, :5].plot(kind='line')

plt.title("Top 5 Skills for Data Analysts")
plt.ylabel("Count")
plt.xlabel("Month")
plt.legend()
plt.show()
```

- clean up this plot by using the month name

```python
df_DA_pivot.reset_index(inplace=True)
df_DA_pivot['job_posted_month'] = (
    pd.to_datetime(df_DA_pivot['job_posted_month_no'], format='%m')
    .dt.strftime('%b')
)
df_DA_pivot.set_index('job_posted_month', inplace=True)
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')

plt.title("Top 5 Skills for Data Analysts")
plt.ylabel("Count")
plt.xlabel("Month")
plt.legend()
plt.show()
```
